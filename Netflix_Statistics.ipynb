SCATTER PLOT

```python
import pandas as pd
import matplotlib.pyplot as plt


# reading the database
data = pd.read_csv("/content/archive (1).zip")

# Scatter plot with day against tip
plt.scatter(data['Subscription Type'], data['Plan Duration'])

# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('Subscription Type')
plt.ylabel('Plan Duration')

plt.show()
```

LINE CHART

```python
import pandas as pd
import matplotlib.pyplot as plt


# reading the database
data = pd.read_csv("/content/archive (1).zip")

# Scatter plot with day against tip
plt.plot(data['Age'])
plt.plot(data['Gender'])

# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('Age')
plt.ylabel('Gender')

plt.show()
```

BAR CHART

```python
import pandas as pd
import matplotlib.pyplot as plt


# reading the database
data = pd.read_csv("/content/archive (1).zip")

# Bar chart with day against tip
plt.bar(data['Country'], data['Monthly Revenue'])

plt.title("Bar Chart")

# Setting the X and Y labels
plt.xlabel('Country')
plt.ylabel('Monthly Revenue')

# Adding the legends
plt.show()
```

HISTOGRAM

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


# reading the database
data = pd.read_csv("/content/archive (1).zip")

sns.histplot(x='Subscription Type', data=data, kde=True, hue='Device')

plt.show()
```
